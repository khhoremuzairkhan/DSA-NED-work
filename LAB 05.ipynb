```python
def bubbleSort(seq):
    for i in range(0,len(seq)-1):
        for j in range(i+1,len(seq)):
            if seq[i]>seq[j]:
                seq[i],seq[j]=seq[j],seq[i]
```

```python
def generateRandomList(size,start,end):
    import random
    seq=[]
    for i in range(0,size):
        seq=seq+[random.randint(start,end)]
    return seq
```

```python
# Best case in Bubble sort
for i in range (4000,32001,4000):
    A=generateRandomList(i,1,100000)
    A.sort()
    print(i)
    %timeit bubbleSort(A)
```

```python
import matplotlib.pyplot as plt
y=[0.167,0.510,1.23,3.27,3.96,7.32,8.51,11.9]
x=[]
for i in range(1000,8001,1000):
    x.append(i)
plt.plot(x,y)
plt.xlabel('Array Size (N)')
plt.ylabel('Running Time (seconds)')
plt.title('Best-Case Bubble Sort')
plt.show()
```

```python
# worst case in Bubble sort
for i in range (4000,32001,4000):
    A=generateRandomList(i,1,100000)
    A.sort(reverse=True)
    print(i)
    %timeit bubbleSort(A)
```

```python
def insertionSort (seq):
    for j in range (1,len(seq)):
        i=j-1
        key=seq[j]
        while (i>=0) and seq[i]>key:
            seq[i+1]=seq[i]
            i=i-1
            seq[i+1]=key
```

```python
# Best case in insertion sort
for i in range (4000,32001,4000):
    A=generateRandomList(i,1,100000)
    A.sort()
    print(i)
    %timeit insertionSort(A)
```

```python
def selectionSort(seq):
    for i in range (len(seq)):
        min_index=i
        for j in range (i+1,len(seq)):
            if seq[min_index]>seq[j]:
                min_index=j
        seq[i],seq[min_index]=seq[min_index],seq[i]
```

```python
# Best case in selection sort
for i in range (4000,32001,4000):
    A=generateRandomList(i,1,100000)
    A.sort()
    print(i)
    %timeit selectionSort(A)
```
