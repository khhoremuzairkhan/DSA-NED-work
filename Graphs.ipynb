```python
#bubble sort best case
import matplotlib.pyplot as plt
y=[0.0495,0.213,0.438,0.794,1.24,1.78,2.46,3.18]
x=[]
for i in range(1000,8001,1000):
    x.append(i)
plt.plot(x,y)
plt.title('Bubble Sort Best Case')
plt.xlabel('size of array (n)')
plt.ylabel('time (s)')
plt.show()
```
